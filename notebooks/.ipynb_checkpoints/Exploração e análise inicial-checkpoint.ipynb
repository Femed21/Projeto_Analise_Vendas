```python
# Carregar os dados processados

import pandas as pd

# Carregar o dataset processado
df = pd.read_csv(r"C:\Users\felip\OneDrive\Desktop\Estudos e Trabalhos\DSP - Data Science Project\Nível 1\Projeto 1 - Ciência de Dados\data\processed\dataset_limpo.csv")

# Exibir as primeiras linhas
df.head()
```

```python
# Verificar estatísticas básicas

df.info() # Tipos de dados e valores nulos
df.describe() # Estatísticas gerais (média, desvio padrão, etc.)
df.nunique() # Contagem de valores únicos por coluna
```

```python
# Exploração de colunas importantes

import matplotlib.pyplot as plt
import seaborn as sns

# Definir o estilo dos gráficos
sns.set_style("whitegrid")

# Histograma dos preços
plt.figure(figsize=(8,5))
sns.histplot(df["preco"], bins=30, kde=True)
plt.title("Distribuição dos Preços")
plt.show()

# Contagem de produtos vendidos por categoria
plt.figure(figsize=(10, 5))
sns.countplot(y=df["categoria"], order=df["categoria"].value_counts().index)
plt.title("Quantidade de Produtos Vendidos por Categoria")
plt.show()

# Contagem de produtos vendidos por loja
plt.figure(figsize=(10, 5))
sns.countplot(y=df["unidade"], order=df["unidade"].value_counts().index)
plt.title("Quantidade de Produtos Vendidos por Unidade")
plt.show()
```

```python
# Análise de tendências

df["data_venda"] = pd.to_datetime(df["data_venda"]) # Converter se necessário
df.set_index("data_venda").resample("M")["quantidade"].sum().plot(figsize=(10, 5))
plt.title("Total de Vendas por Mês")
plt.ylabel("Quantidade Vendida")
plt.show()
```
